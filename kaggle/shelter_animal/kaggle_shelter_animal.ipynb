```python
from datetime import datetime
```

```python
raw_shelter_animal['DateTime'] = raw_shelter_animal['DateTime'].apply(lambda date_time: datetime.strptime(date_time, '%Y-%m-%d %H:%M:%S'))
```

```python
raw_shelter_animal.dropna(inplace=True)
```

```python
def convert_to_days(item):
    day_table = {
        'years' : 365,
        'year'  : 365,
        'months': 30,
        'month' : 30,
        'weeks' : 7,
        'week'  : 7,
        'days'  : 1,
        'day'   : 1
    }
    return int(item[0]) * day_table[str(item[1])]
raw_shelter_animal['AgeuponOutcome'] = raw_shelter_animal['AgeuponOutcome'].str.split(' ').apply(convert_to_days)
```
